Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading Data

```python
df = pd.read_csv('Twitch Dataset.csv')
```

Top 10 Streamers

```python
# Top 10 streamers by average viewers per stream

top_10_viewers = df.nlargest(10, 'AVG_VIEWERS_PER_STREAM')[['NAME', 'AVG_VIEWERS_PER_STREAM']]

# Plotting the top 10 streamers by average viewers per stream
plt.figure(figsize=(12, 6))
sns.barplot(x='AVG_VIEWERS_PER_STREAM', y='NAME', data=top_10_viewers, palette='viridis')
plt.title('Top 10 Streamers by Average Viewers per Stream')
plt.xlabel('Average Viewers per Stream')
plt.ylabel('Streamer Name')
plt.show()

print(top_10_viewers)
```

Language Analysis

```python
# Language analysis: Most common languages among the top streamers

language_counts = df['LANGUAGE'].value_counts().head(10)

# Plotting the most common languages

plt.figure(figsize=(12, 6))
sns.barplot(x=language_counts.values, y=language_counts.index, palette='viridis')
plt.title('Most Common Languages Among Top Streamers')
plt.xlabel('Number of Streamers')
plt.ylabel('Language')
plt.show()

print(language_counts)
```

Game Popularity

```python
# Game popularity: Most frequently streamed games by the top streamers

most_streamed_games = df['MOST_STREAMED_GAME'].value_counts().head(10)

# Plotting the most frequently streamed games

plt.figure(figsize=(12, 6))
sns.barplot(x=most_streamed_games.values, y=most_streamed_games.index, palette='viridis')
plt.title('Most Frequently Streamed Games by Top Streamers')
plt.xlabel('Number of Streamers')
plt.ylabel('Game')
plt.show()

print(most_streamed_games)
```

Followers gained

```python
# Top 10 streamers by followers gained per stream

top_10_followers = df.nlargest(10, 'FOLLOWERS_GAINED_PER_STREAM')[['NAME', 'FOLLOWERS_GAINED_PER_STREAM']]

# Plotting the top 10 streamers by followers gained per stream

plt.figure(figsize=(12, 6))
sns.barplot(x='FOLLOWERS_GAINED_PER_STREAM', y='NAME', data=top_10_followers, palette='viridis')
plt.title('Top 10 Streamers by Followers Gained per Stream')
plt.xlabel('Followers Gained per Stream')
plt.ylabel('Streamer Name')
plt.show()

print(top_10_followers)
```

Most Active Days

```python

# Calculate the most active days for streaming
most_active_days = df['MOST_ACTIVE_DAY'].value_counts().head(7)

# Plotting the most active days for streaming using a pie chart
plt.figure(figsize=(10, 8))
plt.pie(most_active_days, labels=most_active_days.index, autopct='%1.1f%%', colors=sns.color_palette('viridis', len(most_active_days)))
plt.title('Most Active Days for Streaming Among Top Streamers')
plt.show()

print(most_active_days)
```
